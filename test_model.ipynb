```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Load your trained model
model = YOLO('/home/sunny/CCRL_vs/ocr/datasets/runs/detect/yolov8_text_seg2/weights/best.pt')
model.to('cuda:7')
results = model.predict('/home/sunny/CCRL_vs/ocr/sample_page.jpg', conf=0.25)  # Predict image

def show_results(results):
    """Visualize results using matplotlib"""
    for r in results:
        # Convert BGR to RGB and create plot
        im_array = r.plot()  # Get BGR numpy array
        im_rgb = cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB)
        
        # Create matplotlib figure
        plt.figure(figsize=(12, 8))
        plt.imshow(im_rgb)
        plt.axis('off')  # Hide axes
        plt.show()

# Inference on an image
image_results = model('/home/sunny/CCRL_vs/ocr/sample_page.jpg')
show_results(image_results)
```

```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load your trained model
model = YOLO('/home/sunny/CCRL_vs/ocr/datasets/runs/detect/train5/weights/best.pt')
print(model.names)
model.to('cuda:7')

def extract_text_blocks1(image):
    # Run inference
    results = model.predict(
        image,
        imgsz=640,
        conf=0.3,
        iou=0.3,
        device=model.device,
        verbose=False
    )[0]
    
    text_blocks = []
    boxes_info = []
    class_names = model.names
    h, w = image.shape[:2]

    if results.boxes:
        boxes = results.boxes.xyxy.cpu().numpy()
        class_ids = results.boxes.cls.cpu().numpy().astype(int)
        
        # Collect all detected text blocks with coordinates
        for box, cls_id in zip(boxes, class_ids):
            if class_names[cls_id] == "item":  # Verify your class name
                x1, y1, x2, y2 = map(int, box)
                if x2 > x1 and y2 > y1:
                    boxes_info.append({
                        'coords': (x1, y1, x2, y2),
                        'crop': image[y1:y2, x1:x2]
                    })

        # Sort blocks in reading order using grid-based grouping
        if boxes_info:
            # Create vertical grid (adjust 20 based on your document line spacing)
            grid_y = 20
            vertical_grid = np.arange(0, h, grid_y)
            
            # Create horizontal grid (adjust 20 based on column spacing)
            grid_x = 20
            horizontal_grid = np.arange(0, w, grid_x)
            
            # Create sorting keys with grid snapping
            sorted_blocks = []
            for box in boxes_info:
                x1, y1, x2, y2 = box['coords']
                center_y = (y1 + y2) // 2
                center_x = (x1 + x2) // 2
                
                # Snap to nearest grid lines
                snap_y = min(vertical_grid, key=lambda y: abs(y - center_y))
                snap_x = min(horizontal_grid, key=lambda x: abs(x - center_x))
                
                # Create sorting key (prioritize vertical then horizontal)
                sorted_blocks.append((
                    snap_y,  # Primary sort by vertical position
                    snap_x,  # Secondary sort by horizontal position
                    box['crop']  # Keep reference to the crop
                ))

            # Sort blocks using the grid-based keys
            sorted_blocks.sort(key=lambda x: (x[0], x[1]))
            
            # Extract sorted crops
            text_blocks = [block[2] for block in sorted_blocks]

    return text_blocks

def extract_text_blocks(image):
    # Handle large images
    h, w = image.shape[:2]
    if max(h, w) > 1280:
        scale = 1280 / max(h, w)
        image = cv2.resize(image, (int(w*scale), int(h*scale)))
    
    # Run inference
    results = model.predict(
        image,
        imgsz=640,
        conf=0.4,  # Optimal balance for document text
        iou=0.4,
        device=model.device,
        verbose=False
    )[0]
    
    # Process results
    text_blocks = []
    class_names = model.names
    print(class_names)
    
    if results.boxes:
        boxes = results.boxes.xyxy.cpu().numpy()
        class_ids = results.boxes.cls.cpu().numpy().astype(int)
        # print(boxes, class_ids)
        
        for box, cls_id in zip(boxes, class_ids):
            # print(class_names[cls_id])
            if class_names[cls_id] == "item":
                x1, y1, x2, y2 = map(int, box)
                # Safety check for valid coordinates
                if x2 > x1 and y2 > y1:
                    text_blocks.append(image[y1:y2, x1:x2])
    
    return text_blocks

def show_results(results):
    count = 0
    """Visualize results using matplotlib"""
    for r in results:
        count += 1
        # Convert BGR to RGB and create plot
        im_array = r.plot()  # Get BGR numpy array
        im_rgb = cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB)
        
        # Create matplotlib figure
        plt.figure(figsize=(12, 8))
        plt.imshow(im_rgb)
        plt.axis('off')  # Hide axes
        plt.savefig('text_boxes.png')

def nearest_in_list(number, list_of_numbers):
    closest = min(list_of_numbers, key=lambda x:abs(x-number))
    if abs(closest - number) == abs(number - min(n for n in list_of_numbers if n > number)):
        return min(n for n in list_of_numbers if n > number)
    else:
        return closest
    
image_results = model('/home/sunny/CCRL_vs/ocr/text-detection-positive.png', verbose=False, conf=0.25, iou=0.35, half=True, augment=True)  # Predict image
image = cv2.imread('/home/sunny/CCRL_vs/ocr/text-detection-positive.png')
print(image.shape)
show_results(image_results)

result = model(np.copy(image))[0] # detect

cuts = []
boxes = []
for box in result.boxes:
    box = box.xyxy
    x, y, x2, y2 = int(box[0, 0].item()), int(box[0, 1].item()), int(box[0, 2].item()), int(box[0, 3].item())
    cuts.append(image[y:y2, x:x2, :])
    boxes.append((x, y, x2, y2))
    image = cv2.rectangle(image, (x, y), (x2, y2), (0, 0, 0), 1)

plt.imshow(image)

## organize words
xs = {}
for i in range(len(cuts)):
    x = nearest_in_list(boxes[i][1], np.arange(0, image.shape[0], 22).tolist())
    y = nearest_in_list(boxes[i][0], np.arange(0, image.shape[1], 22).tolist())

    xs[str(i)] = x + y*0.01
xs = sorted(xs.items(), key=lambda x:x[1], reverse=False)

for index, val in xs:
    plt.figure(figsize=(1, 1))
    plt.imshow(cuts[int(index)])


# output = extract_text_blocks(image)
# print(len(output))

# for i, block in enumerate(output):
#     plt.figure(figsize=(3, 3))
#     plt.imshow(cv2.cvtColor(block, cv2.COLOR_BGR2RGB))
#     plt.title(f"Block {i+1}")
#     plt.axis('off')
# plt.show()

# output1 = extract_text_blocks1(image)
# print(len(output1))

# for i, block in enumerate(output1):
#     plt.figure(figsize=(3, 3))
#     plt.imshow(cv2.cvtColor(block, cv2.COLOR_BGR2RGB))
#     plt.title(f"Block {i+1}")
#     plt.axis('off')
# plt.show()
```
